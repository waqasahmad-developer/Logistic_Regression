# Perceptron-trick

```python
from sklearn.datasets import make_classification
import numpy as np
X,y = make_classification(n_samples=100, n_features=2, n_informative=1, n_redundant=0, n_classes=2, n_clusters_per_class=1,random_state=41, hypercube=False, class_sep=10)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y, cmap='winter', s=100)
```

```python
def perceptron(X,y):

  X = np.insert(X,0,1,axis=1)
  weights = np.zeros(X.shape[1])
  lr = 0.1

  for i in range(1000):
    j = np.random.randint(0,100)         # selecting a random row
    y_hat = step(np.dot(X[j], weights))  # this step function returns 1 if number>0 else zero
    weights = weights +lr*(y[j] - y_hat)*X[j]   # updating the weights

  return weights[0], weights[1:]      # weights [0]  -> intercept   weights [1:]   -> coef
```

```python
# defining the step function
def step(z):
  return 1 if z > 0 else 0
```

```python
intercept_, coef_ = perceptron(X,y)
```

```python
print(coef_)
print(intercept_)
```

```python
# m = -(A/B)    and   b = -(C/B)
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + b
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input, y_input, color='red', linewidth=3)
plt.scatter(X[:,0],X[:,1], c=y, cmap='winter', s=100)
plt.ylim(-3,2)
```

```python
def perceptron(X,y):

  m = []
  b = []

  X = np.insert(X,0,1, axis=1)
  weights  = np.zeros(X.shape[1])
  lr = 0.1

  for i in range(200):
    j = np.random.randint(0,100)
    y_hat = step(np.dot(X[j],weights))
    weights = weights +lr*(y[j] - y_hat) * X[j]

    m.append(-(weights[1] / weights[2]))
    b.append(-(weights[0] / weights[2]))
  return m,b
```

```python
m,b = perceptron(X,y)
```
